# Superstore Sales Dataset Cleaning and Preprocessing

```python
import pandas as pd
```

```python
df = pd.read_csv("SuperStoreOrders.csv")
df
```

```python
df.info()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='mixed', dayfirst=True, errors='coerce')
print("Unparsed dates:", df['order_date'].isna().sum())
```

```python
df['ship_date'] = pd.to_datetime(df['ship_date'], format='mixed', dayfirst=True, errors='coerce')
print("Unparsed dates:", df['ship_date'].isna().sum())
```

```python
df['sales'] = df['sales'].str.replace(',', '', regex=False).astype(float)
```

```python
df.to_csv("Cleaned_Superstore.csv")
```
